```python
from google import genai
from dotenv import load_dotenv
from IPython.display import Markdown, Image
from google.genai import types
import os
```

```python
# Load .env
load_dotenv()

API_KEY = os.getenv("GEMINI_API_KEY")
MODEL_ID = os.getenv("GEMINI_MODEL_ID")
IMAGE_MODEL_ID = os.getenv("GEMINI_IMAGE_MODEL_ID")

client = genai.Client(api_key=API_KEY)
```

```python
chat = client.chats.create(model=MODEL_ID)

res = chat.send_message("In the scene bridge of death in motty python, what if the knights couldn't answer the questions?")
Markdown(res.text)
```

```python
res = chat.send_message("Can you tell me, how did it go?")
Markdown(res.text)
```

```python
# seeing the message history
for message in chat.get_history():
    print(f'role - {message.role}',end=": ")
    print(message.parts[0].text)
```

```python
res = client.models.generate_content(
    model=IMAGE_MODEL_ID,
    contents='Create the aftermath scene of the bridge of death in motty python above',
    config=types.GenerateContentConfig(
        response_modalities=['Text', 'Image']
    )
)

for part in res.candidates[0].content.parts:
  if part.text is not None:
    display(Markdown(part.text))
  elif part.inline_data is not None:
    mime = part.inline_data.mime_type
    print(mime)
    data = part.inline_data.data
    display(Image(data=data))
```
