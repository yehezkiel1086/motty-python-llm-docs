```python
from google import genai
from dotenv import load_dotenv
from IPython.display import Markdown
import os
```

```python
# Load .env
load_dotenv()

API_KEY = os.getenv("GEMINI_API_KEY")
MODEL_ID = os.getenv("GEMINI_MODEL_ID")

client = genai.Client(api_key=API_KEY)
```

### Simple Prompt

```python
res = client.models.generate_content(
  model=MODEL_ID,
  contents=[
    'What is the capital of Assyria?',
  ]
)

res.text
```

```python
Markdown(res.text)
```

### Stream Prompt

```python
res = client.models.generate_content_stream(
  model=MODEL_ID,
  contents="Explain how Arthur became King in Motty Python"
)

for chunk in res:
  print(chunk.text)
  print("_" * 100)
```
