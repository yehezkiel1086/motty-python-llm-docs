```python
from google import genai
from dotenv import load_dotenv
from IPython.display import Markdown
import os
from PIL import Image
import base64
```

```python
# Load .env
load_dotenv()

API_KEY = os.getenv("GEMINI_API_KEY")
MODEL_ID = os.getenv("GEMINI_MODEL_ID")

client = genai.Client(api_key=API_KEY)
```

### Extract text from Image

```python
img = Image.open("asset/35-Best-Monty-Python-Memes-That-Will-Make-You-Say-A-Blessing-From-The-Lord-FG.webp")
img.show() # BEWARE: jumpscare
```

```python
with open("asset/35-Best-Monty-Python-Memes-That-Will-Make-You-Say-A-Blessing-From-The-Lord-FG.webp", "rb") as f:
    image_data = f.read()
prompt = "Explain the meme to an old lady" # enter prompt

res = client.models.generate_content(
    model=MODEL_ID,
    contents=[
      {"role": "user", "parts": [
        {"text": prompt},
        {"inline_data": {
            "mime_type": "image/webp",
            "data": base64.b64encode(image_data).decode("utf-8")
        }}
      ]}
    ]
)

print(res.text)
```
